```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read Iris.csv file-->>
df = pd.read_csv('/content/Iris (1).csv')
df
```

```python
# identify information about all columns
df.info()
```

```python
# create histogram of sepal width
plt.hist(df['SepalWidthCm'], edgecolor= 'k' , color = 'b', alpha =0.5)
plt.title('Histogram of Sepal Width', fontsize = 20, color = 'red')
plt.xlabel('Sepal Width in CM')
plt.ylabel('frequency')
plt.show()
```

```python
# create histogram of sepal length
plt.hist(df['SepalLengthCm'], edgecolor = 'k', color = 'y')
plt.title('Histogram of Sepal Length', color ='red', fontsize =20)
plt.xlabel('Sepal Length in cm')
plt.ylabel('frequency')
plt.show()
```

```python
# create histogram of petal width
plt.hist(df['PetalWidthCm'], edgecolor ='k', color ='grey')
plt.title('Histogram of Petal Width', color = 'red', fontsize = 20)
plt.xlabel('Petal Width in cm')
plt.ylabel('frequency')
plt.show()
```

```python
# create histogram of petal length
plt.hist(df['PetalLengthCm'], edgecolor ='k', color ='orange')
plt.title('Histogram of Petal Length', color = 'red', fontsize = 20)
plt.xlabel('Petal Length in cm')
plt.ylabel('frequency')
plt.show()
```

```python
# create histogram of Species
species_count = df['Species'].value_counts()
plt.bar(species_count.index, species_count.values, color ='tan', edgecolor ='k')
plt.title('Distrubution', color = 'red', fontsize = 20)
plt.xlabel('Species')
plt.ylabel('frequency')
plt.show()
```
